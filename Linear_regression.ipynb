```python
import pandas as pd
```

```python
Data = pd.read_excel(r'/content/gdrive/My Drive/Height and Width.xlsx')
```

```python
display(pd.DataFrame(Data))
```

```python
Data.rent[0:6]
```

```python
x=Data[['Height']]
```

```python
x.head(6)
```

```python
y=Data[["Weight"]]
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
xtrain,xtest,ytrain,ytest=tts(x,y,test_size=0.3,random_state=5)
```

```python
xtrain.head()
```

```python
xtest.head()
```

```python
ytrain.head()
```

```python
ytest.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
N=LinearRegression()
```

```python
N.fit(xtrain,ytrain)
```

```python
M=N.coef_#y=mx+c
```

```python
M
```

```python
c=N.intercept_
```

```python
N.predict(xtest)
```

```python
Data["predicted Result"]=N.predict(x)
```

```python
Data.head()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
pred=N.predict(xtest)
```

```python
pred
```

```python
mse=mean_squared_error(ytest,pred)
```

```python
mse
```

```python
mae=mean_absolute_error(ytest,pred)
```

```python
mae
```

```python
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(50,35))
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height_Width_Plot")
plt.scatter(xtrain,ytrain)
plt.plot(xtrain,N.predict(xtrain))
plt.scatter(xtrain,N.predict(xtrain),color="green")
```
